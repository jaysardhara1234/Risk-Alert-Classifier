# Part A : Conceptual Understanding

## Q1. What is Logistic Regression and why is it suitable for classification ?

### Logistic Regression is a supervised machine learning algorithm used for binary classification problems. It predicts the probability that an observation belongs to a particular class using the sigmoid function, which produces values between 0 and 1. A threshold (commonly 0.5) is then used to classify the observation into one of the two classes.

## Q2. Explain classification performance metrics and why accuracy alone is insufficient.

### Classification performance is evaluated using metrics such as Accuracy, Precision, Recall, F1-score, and ROC-AUC. Accuracy measures the overall proportion of correct predictions but can be misleading for imbalanced datasets. Metrics like Precision, Recall, and F1-score provide a better assessment of how well the model identifies the minority class.

## Q3. Define Type-I Error and Type-II Error in the context of risk prediction.

### Type-I Error (False Positive): Predicting a customer as high risk when they are actually low risk, which may lead to rejecting a trustworthy customer.

### Type-II Error (False Negative): Predicting a customer as low risk when they are actually high risk, which may result in approving a risky customer and increasing financial losses.

## Q4. Explain Precision, Recall, F1-Score, TPR, and FPR.

### Precision: Measures the proportion of predicted positive cases that are actually positive.
### Recall (TPR): Measures the proportion of actual positive cases correctly identified by the model.
### F1-Score: The harmonic mean of Precision and Recall, providing a balanced evaluation.
### True Positive Rate (TPR): Another name for Recall.
### False Positive Rate (FPR): Measures the proportion of actual negative cases incorrectly classified as positive.

## Q5. What is AUC-ROC and how does it help in evaluating classifiers ?

### ROC (Receiver Operating Characteristic) Curve plots the True Positive Rate against the False Positive Rate at different classification thresholds. The Area Under the Curve (AUC) summarizes the model's ability to distinguish between positive and negative classes. A higher AUC indicates better class separation, with 1.0 representing a perfect classifier.

## Q6. Why does imbalanced data create problems in classification models ?

### Imbalanced data occurs when one class has significantly more observations than the other. Models trained on such data tend to favor the majority class, resulting in poor detection of the minority class despite achieving high accuracy. Therefore, metrics such as Precision, Recall, and F1-score are preferred over accuracy for evaluating performance on imbalanced datasets.
